```python
import pandas as pd
df = pd.read_csv('instagram_Analytics.csv')
df.head()
```

```python
df = df.drop(['post_id','upload_date'],axis="columns")
df.head()
```

```python
df.info()
```

```python
df = pd.get_dummies(
    df,
    columns=['media_type', 'traffic_source', 'content_category'],
    drop_first=True,
    dtype=int
)
df.head()
```

```python
x = df.drop(['engagement_rate'],axis="columns")
x.head()
```

```python
y = df['engagement_rate']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.1, random_state = 42)
# here the data was splitted ito two sets by using the test train split
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
predicted = model.predict(X_test)
# The model uses X_test to make predictions internally,
# then compares them to y_test.
result = model.score(X_test, y_test)

print(f"R2 Score: {result:.4f}")
```
