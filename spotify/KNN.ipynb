```python
import pandas as pd
df = pd.read_csv('spotify.csv')
df.head()
```

```python
df['hit'] = df['track_popularity'].apply(lambda x: 1 if x >= 50 else 0)
df = df.drop(['track_popularity','track_name'], axis="columns")
df.head()
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['hit'],axis="columns"), df['hit'], test_size = 0.1, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
# fine tuning with the values
model = RandomForestClassifier(n_estimators = 400,max_depth=10,min_samples_split=5)
model.fit(X_train,y_train)
model.fit(X_train, y_train)
```

```python
# predicting the values
y_pred = model.predict(X_test)
```

```python
# Getting the probability of the model

print(f"Accuracy {model.score(X_test, y_test)}")
```
