```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Load the data (assuming you have saved it as 'Mall_Customers.csv')
# Note: You may need to replace 'Mall_Customers.csv' with the correct path/filename.
try:
    df = pd.read_excel("OnlineRetail.xlsx")
except FileNotFoundError:
    print("Error: 'Mall_Customers.csv' not found. Please ensure the file is in the correct directory.")
    exit()

# We focus on the two most relevant features for visualization:
# Annual Income (k$) and Spending Score (1-100)
X = df.iloc[:, [3, 4]].values

print("✅ Data loaded and features selected.")

# Initial Scatter Plot (before clustering)
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, c='gray', alpha=0.7)
plt.title('Mall Customer Data (Raw)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
```
