```python
import pandas as pd
df = pd.read_csv('spotify.csv')
df.head()
```

```python
df['hit'] = df['track_popularity'].apply(lambda x: 1 if x >= 50 else 0)
df = df.drop(['track_popularity','track_name'], axis="columns")
df.head()
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['hit'],axis="columns"), df['hit'], test_size = 0.1, random_state = 3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500,C=1)
model.fit(X_train, y_train)
```

```python
# predicting the values
y_pred = model.predict(X_test)
```

```python
# Getting the probability of the model
from sklearn.metrics import confusion_matrix

print(f"Accuracy {model.score(X_test, y_test)}")

Y_pred = model.predict(X_test)
con_mat = confusion_matrix(y_test, Y_pred)
print("The confusion matrix")
print(con_mat)
```
