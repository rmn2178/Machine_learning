```python
import sklearn.svm
# importing the essential libraries
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
from sklearn.svm import SVC

from iris_train.supp_vect_mach import X_train
```

```python
# Splitting the data into test and train
digits = load_digits()
X_train,X_test,Y_train,Y_test = train_test_split(digits.data,digits.target,test_size=0.1,random_state=0)
len(X_train)
```

```python
# training using the logistic regression
model_x = LogisticRegression(max_iter=1000)
model_x.fit(X_train,Y_train)
model_x.score(X_test,Y_test)
```

```python
# training using the SVC
model_y = SVC(max_iter=1000)
model_y.fit(X_train,Y_train)
model_y.score(X_test,Y_test)
```

```python
# training using the RandomForestClassifier
model_z = RandomForestClassifier()
model_z.fit(X_train,Y_train)
model_z.score(X_test,Y_test)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5)
```

```python
from sklearn.model_selection import cross_val_score
# for the logistic regression model
cross_val_score(model_x,X_train,Y_train)
```

```python
# here for the svm
cross_val_score(model_y,X_train,Y_train)
```

```python
# here for the RFA
cross_val_score(model_z,X_train,Y_train)
```
