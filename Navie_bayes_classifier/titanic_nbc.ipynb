```python
# reading the data
import pandas as pd
df = pd.read_csv("titanic_data.csv")
df.head()
```

```python
# we doesnt need all the columns
df.drop(['Name','SibSp','Parch','PassengerId','Ticket','Cabin','Embarked'], axis=1, inplace=True)
df.head()
```

```python
# making data separately
target = df.Survived
inputs = df.drop('Survived', axis='columns')
inputs.head()
```

```python
dumy = pd.get_dummies(inputs.Sex).astype(int)
dumy.head()
```

```python
# here dropping the Sex and adding the dumy
inputs.drop('Sex', axis="columns", inplace=True)
inputs = pd.concat([inputs, dumy], axis="columns")
inputs.head()
```

```python
# preprocessing the data
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

Here we can note that the values of the ade is sometimes nan which means that the data of the age column is not entered sometimes

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
# here we are splitting the datasets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs,target,test_size=0.1,random_state=42)
x_train.shape
```

```python
# here we are training
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
 # printing the y_test
model.predict_proba(x_test[:10])
```
